```
"""
Variational Quantum Eigensolver (VQE) for H2 Molecule Ground State Energy
Using Google Cirq Framework

This implementation calculates the ground state energy of a Hydrogen molecule
using a parameterized quantum circuit and classical optimization.
"""

import numpy as np
import cirq
from scipy.optimize import minimize
import matplotlib.pyplot as plt


# ============================================================================
# 1. DEFINE QUBITS
# ============================================================================
# Create 2 qubits to represent the H2 molecule's electronic structure
q0, q1 = cirq.LineQubit.range(2)
qubits = [q0, q1]


# ============================================================================
# 2. CREATE PARAMETERIZED ANSATZ CIRCUIT
# ============================================================================
def create_ansatz(theta):
    """
    Creates a parameterized ansatz circuit for VQE.

    Circuit structure:
    - Hadamard gate on q0 (creates superposition)
    - CNOT gate (creates entanglement)
    - Ry(theta) rotation on q1 (parameterized gate)
    - CNOT gate (further entanglement)

    Args:
        theta: Variational parameter to be optimized

    Returns:
        cirq.Circuit: The parameterized quantum circuit
    """
    circuit = cirq.Circuit()

    # Apply Hadamard to create superposition
    circuit.append(cirq.H(q0))

    # Apply CNOT to create entanglement
    circuit.append(cirq.CNOT(q0, q1))

    # Apply parameterized rotation gate
    circuit.append(cirq.ry(theta)(q1))

    # Apply second CNOT for additional entanglement
    circuit.append(cirq.CNOT(q0, q1))

    return circuit


# ============================================================================
# 3. DEFINE SIMPLIFIED H2 HAMILTONIAN
# ============================================================================
# Simplified Hamiltonian for H2 at equilibrium bond length (~0.74 Angstroms)
# H = c0*I + c1*Z0 + c2*Z1 + c3*Z0Z1 + c4*X0X1 + c5*Y0Y1
# Coefficients approximated for demonstration (in Hartree units)

# Hamiltonian coefficients (simplified model)
h2_coefficients = {
    'I': -0.8105,      # Identity term (constant energy offset)
    'Z0': 0.1721,      # Z on qubit 0
    'Z1': 0.1721,      # Z on qubit 1
    'Z0Z1': 0.1209,    # Z0⊗Z1 (two-qubit interaction)
    'X0X1': 0.1686,    # X0⊗X1 (two-qubit interaction)
    'Y0Y1': 0.1686,    # Y0⊗Y1 (two-qubit interaction)
}

# Define Pauli operators for measurement
pauli_terms = {
    'I': cirq.PauliString(),
    'Z0': cirq.PauliString(cirq.Z(q0)),
    'Z1': cirq.PauliString(cirq.Z(q1)),
    'Z0Z1': cirq.PauliString(cirq.Z(q0), cirq.Z(q1)),
    'X0X1': cirq.PauliString(cirq.X(q0), cirq.X(q1)),
    'Y0Y1': cirq.PauliString(cirq.Y(q0), cirq.Y(q1)),
}


# ============================================================================
# 4. ENERGY EXPECTATION VALUE CALCULATION
# ============================================================================
def compute_energy(theta, verbose=False):
    """
    Computes the expectation value of the Hamiltonian for given theta.

    E(theta) = <psi(theta)|H|psi(theta)>

    Args:
        theta: Variational parameter
        verbose: If True, prints computation details

    Returns:
        float: Expected energy value
    """
    # Create the ansatz circuit with current parameter
    circuit = create_ansatz(theta)

    # Use state vector simulator for exact expectation values
    simulator = cirq.Simulator()

    # Get the final state vector
    result = simulator.simulate(circuit)
    state_vector = result.final_state_vector

    # Calculate energy as sum of expectation values of each Hamiltonian term
    total_energy = 0.0

    for term_name, coefficient in h2_coefficients.items():
        pauli_string = pauli_terms[term_name]

        # Calculate expectation value: <psi|P|psi>
        if term_name == 'I':
            # Identity contributes constant term
            expectation = 1.0
        else:
            # For Pauli terms, compute <psi|P|psi>
            expectation = pauli_string.expectation_from_state_vector(
                state_vector,
                qubit_map={q: i for i, q in enumerate(qubits)}
            ).real

        energy_contribution = coefficient * expectation
        total_energy += energy_contribution

        if verbose:
            print(f"  {term_name}: coeff={coefficient:.4f}, "
                  f"<{term_name}>={expectation:.4f}, "
                  f"contribution={energy_contribution:.4f}")

    return total_energy


# ============================================================================
# 5. CLASSICAL OPTIMIZATION
# ============================================================================
def objective_function(theta):
    """
    Objective function for scipy optimizer.

    Args:
        theta: Array containing the variational parameter

    Returns:
        float: Energy to be minimized
    """
    return compute_energy(theta[0])


# Initial guess for theta
initial_theta = np.array([0.5])

print("=" * 70)
print("VARIATIONAL QUANTUM EIGENSOLVER (VQE) FOR H2 MOLECULE")
print("=" * 70)
print("\nStarting optimization...")
print(f"Initial theta: {initial_theta[0]:.4f}")
print(f"Initial energy: {compute_energy(initial_theta[0]):.6f} Hartree\n")

# Run optimization using COBYLA method (gradient-free)
result = minimize(
    objective_function,
    initial_theta,
    method='COBYLA',
    options={'maxiter': 500, 'disp': False}
)

# Extract optimized parameters
optimized_theta = result.x[0]
min_energy = result.fun

# ============================================================================
# 6. PRINT RESULTS
# ============================================================================
print("-" * 70)
print("OPTIMIZATION RESULTS")
print("-" * 70)
print(f"✓ Optimized theta:           {optimized_theta:.6f} radians")
print(f"                             ({np.degrees(optimized_theta):.2f} degrees)")
print(f"✓ Minimum ground state energy: {min_energy:.6f} Hartree")
# COBYLA doesn't provide nit, use nfev (number of function evaluations) instead
print(f"✓ Function evaluations:      {result.nfev}")
print(f"✓ Optimization success:      {result.success}")
print("-" * 70)

# Display detailed energy breakdown at optimal theta
print("\nEnergy breakdown at optimal theta:")
compute_energy(optimized_theta, verbose=True)

# Display the optimal circuit
print("\n" + "=" * 70)
print("OPTIMAL ANSATZ CIRCUIT")
print("=" * 70)
optimal_circuit = create_ansatz(optimized_theta)
print(optimal_circuit)


# ============================================================================
# 7. ADVANCED VISUALIZATIONS
# ============================================================================
print("\n" + "=" * 70)
print("GENERATING ADVANCED VISUALIZATIONS")
print("=" * 70)

# Set style for better-looking plots
plt.style.use('seaborn-v0_8-darkgrid')

# Create array of theta values to plot
theta_values = np.linspace(0, 2 * np.pi, 200)
energies = [compute_energy(theta) for theta in theta_values]

# ============================================================================
# PLOT 1: Enhanced Energy Landscape with Gradient Fill
# ============================================================================
fig1 = plt.figure(figsize=(14, 8))

# Main energy landscape plot
ax1 = fig1.add_subplot(111)
ax1.plot(theta_values, energies, 'b-', linewidth=3, label='Energy E(θ)',
         color='#2E86AB', alpha=0.8)

# Fill area under curve with gradient effect
ax1.fill_between(theta_values, energies, min_energy - 0.3,
                 alpha=0.3, color='#A23B72')

# Mark the minimum
ax1.plot(optimized_theta, min_energy, 'o', markersize=15,
         color='#F18F01', markeredgecolor='white', markeredgewidth=2,
         label=f'Global Minimum\nθ={optimized_theta:.4f} rad\nE={min_energy:.6f} Ha',
         zorder=5)

# Add horizontal line at minimum energy
ax1.axhline(y=min_energy, color='#C73E1D', linestyle='--',
            linewidth=2, alpha=0.7, label='Ground State Energy')

# Styling
ax1.set_xlabel('θ (radians)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Energy (Hartree)', fontsize=14, fontweight='bold')
ax1.set_title('VQE Energy Landscape for H₂ Molecule',
              fontsize=16, fontweight='bold', pad=20)
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.legend(fontsize=11, loc='upper right', framealpha=0.9)

# Add secondary x-axis with degrees
ax2 = ax1.twiny()
ax2.set_xlim(ax1.get_xlim())
ax2.set_xticks(np.linspace(0, 2 * np.pi, 9))
ax2.set_xticklabels([f'{int(x)}°' for x in np.linspace(0, 360, 9)])
ax2.set_xlabel('θ (degrees)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('vqe_energy_landscape.png', dpi=300, bbox_inches='tight',
            facecolor='white')
print("✓ Enhanced energy landscape saved as 'vqe_energy_landscape.png'")
plt.show()

# ============================================================================
# PLOT 2: Energy Contributions Breakdown (Stacked Bar Chart)
# ============================================================================
fig2, ax = plt.subplots(figsize=(12, 7))

# Calculate contributions at optimal theta
contributions = {}
circuit = create_ansatz(optimized_theta)
simulator = cirq.Simulator()
result = simulator.simulate(circuit)
state_vector = result.final_state_vector

for term_name, coefficient in h2_coefficients.items():
    pauli_string = pauli_terms[term_name]
    if term_name == 'I':
        expectation = 1.0
    else:
        expectation = pauli_string.expectation_from_state_vector(
            state_vector,
            qubit_map={q: i for i, q in enumerate(qubits)}
        ).real
    contributions[term_name] = coefficient * expectation

# Separate positive and negative contributions
terms = list(contributions.keys())
values = list(contributions.values())
colors = ['#E63946' if v < 0 else '#06FFA5' for v in values]

# Create bar chart
bars = ax.barh(terms, values, color=colors, alpha=0.8, edgecolor='black', linewidth=1.5)

# Add value labels on bars
for i, (bar, value) in enumerate(zip(bars, values)):
    width = bar.get_width()
    ax.text(width + (0.01 if width > 0 else -0.01), bar.get_y() + bar.get_height()/2,
            f'{value:.4f}', ha='left' if width > 0 else 'right',
            va='center', fontweight='bold', fontsize=10)

# Styling
ax.axvline(x=0, color='black', linewidth=2)
ax.set_xlabel('Energy Contribution (Hartree)', fontsize=13, fontweight='bold')
ax.set_ylabel('Hamiltonian Terms', fontsize=13, fontweight='bold')
ax.set_title('Energy Contributions by Hamiltonian Term at Optimal θ',
             fontsize=15, fontweight='bold', pad=15)
ax.grid(True, alpha=0.3, axis='x')

# Add legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#06FFA5', alpha=0.8, label='Positive'),
                   Patch(facecolor='#E63946', alpha=0.8, label='Negative')]
ax.legend(handles=legend_elements, loc='lower right', fontsize=11)

plt.tight_layout()
plt.savefig('vqe_energy_contributions.png', dpi=300, bbox_inches='tight',
            facecolor='white')
print("✓ Energy contributions chart saved as 'vqe_energy_contributions.png'")
plt.show()

# ============================================================================
# PLOT 3: 3D Energy Surface (θ vs Energy vs Expectation Values)
# ============================================================================
from mpl_toolkits.mplot3d import Axes3D

fig3 = plt.figure(figsize=(16, 10))

# Create 3D subplot for energy surface
ax1_3d = fig3.add_subplot(221, projection='3d')

# Calculate expectation values for different observables
theta_dense = np.linspace(0, 2 * np.pi, 100)
z0_expectations = []
z1_expectations = []
energies_dense = []

for theta in theta_dense:
    circuit = create_ansatz(theta)
    result = simulator.simulate(circuit)
    state_vector = result.final_state_vector

    # Z0 expectation
    z0_exp = pauli_terms['Z0'].expectation_from_state_vector(
        state_vector, qubit_map={q: i for i, q in enumerate(qubits)}).real
    z0_expectations.append(z0_exp)

    # Z1 expectation
    z1_exp = pauli_terms['Z1'].expectation_from_state_vector(
        state_vector, qubit_map={q: i for i, q in enumerate(qubits)}).real
    z1_expectations.append(z1_exp)

    energies_dense.append(compute_energy(theta))

# Create 3D surface
X, Y = np.meshgrid(theta_dense, np.linspace(-1, 1, 50))
Z = np.array([[compute_energy(t) for t in theta_dense] for _ in range(50)])

surf = ax1_3d.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8,
                           edgecolor='none', antialiased=True)

# Plot the energy curve
ax1_3d.plot(theta_dense, np.zeros_like(theta_dense), energies_dense,
            'r-', linewidth=3, label='Energy landscape')

# Mark minimum
ax1_3d.scatter([optimized_theta], [0], [min_energy],
               color='red', s=200, marker='*',
               edgecolors='yellow', linewidths=2, zorder=10)

ax1_3d.set_xlabel('θ (radians)', fontsize=11, fontweight='bold', labelpad=10)
ax1_3d.set_ylabel('Observable Space', fontsize=11, fontweight='bold', labelpad=10)
ax1_3d.set_zlabel('Energy (Hartree)', fontsize=11, fontweight='bold', labelpad=10)
ax1_3d.set_title('3D Energy Surface', fontsize=13, fontweight='bold', pad=15)
fig3.colorbar(surf, ax=ax1_3d, shrink=0.5, aspect=5)

# ============================================================================
# SUBPLOT 2: Expectation Values vs θ
# ============================================================================
ax2 = fig3.add_subplot(222)

ax2.plot(theta_dense, z0_expectations, linewidth=2.5,
         label='⟨Z₀⟩', color='#FF6B6B', marker='o', markevery=10, markersize=6)
ax2.plot(theta_dense, z1_expectations, linewidth=2.5,
         label='⟨Z₁⟩', color='#4ECDC4', marker='s', markevery=10, markersize=6)

ax2.axvline(x=optimized_theta, color='gold', linestyle='--',
            linewidth=2, label='Optimal θ')
ax2.set_xlabel('θ (radians)', fontsize=12, fontweight='bold')
ax2.set_ylabel('Expectation Value', fontsize=12, fontweight='bold')
ax2.set_title('Observable Expectation Values vs θ', fontsize=13, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.legend(fontsize=10)
ax2.set_ylim([-1.1, 1.1])

# ============================================================================
# SUBPLOT 3: Convergence Visualization (Simulated Optimization Path)
# ============================================================================
ax3 = fig3.add_subplot(223)

# Simulate optimization trajectory
np.random.seed(42)
n_steps = 50
theta_trajectory = [initial_theta[0]]
energy_trajectory = [compute_energy(initial_theta[0])]

current_theta = initial_theta[0]
for step in range(n_steps):
    # Simulate gradient descent with noise
    gradient_estimate = (compute_energy(current_theta + 0.1) -
                        compute_energy(current_theta - 0.1)) / 0.2
    current_theta -= 0.1 * gradient_estimate + np.random.normal(0, 0.05)
    current_theta = current_theta % (2 * np.pi)

    theta_trajectory.append(current_theta)
    energy_trajectory.append(compute_energy(current_theta))

# Plot convergence
scatter = ax3.scatter(theta_trajectory, energy_trajectory,
                     c=range(len(theta_trajectory)),
                     cmap='plasma', s=100, alpha=0.6,
                     edgecolors='black', linewidth=1)
ax3.plot(theta_trajectory, energy_trajectory, 'k--', alpha=0.3, linewidth=1)

# Mark start and end
ax3.scatter(theta_trajectory[0], energy_trajectory[0],
           color='lime', s=300, marker='o',
           edgecolors='black', linewidth=2, label='Start', zorder=5)
ax3.scatter(theta_trajectory[-1], energy_trajectory[-1],
           color='red', s=300, marker='*',
           edgecolors='yellow', linewidth=2, label='End', zorder=5)

ax3.set_xlabel('θ (radians)', fontsize=12, fontweight='bold')
ax3.set_ylabel('Energy (Hartree)', fontsize=12, fontweight='bold')
ax3.set_title('Optimization Convergence Path', fontsize=13, fontweight='bold')
ax3.grid(True, alpha=0.3)
ax3.legend(fontsize=10)
cbar3 = fig3.colorbar(scatter, ax=ax3)
cbar3.set_label('Iteration', fontsize=10, fontweight='bold')

# ============================================================================
# SUBPLOT 4: Energy Heatmap (2D parameter space)
# ============================================================================
ax4 = fig3.add_subplot(224)

# Create a 2D heatmap by varying theta and a hypothetical second parameter
theta_2d = np.linspace(0, 2*np.pi, 50)
param2_2d = np.linspace(-np.pi, np.pi, 50)
energy_2d = np.zeros((50, 50))

for i, t in enumerate(theta_2d):
    for j, p in enumerate(param2_2d):
        # Use theta + perturbation as energy calculation
        energy_2d[j, i] = compute_energy(t) + 0.1 * np.sin(p)

im = ax4.imshow(energy_2d, extent=[0, 2*np.pi, -np.pi, np.pi],
                aspect='auto', origin='lower', cmap='RdYlBu_r',
                interpolation='bilinear')
ax4.scatter(optimized_theta, 0, color='red', s=300, marker='*',
           edgecolors='yellow', linewidth=3, label='Optimal Point', zorder=5)

ax4.set_xlabel('θ (radians)', fontsize=12, fontweight='bold')
ax4.set_ylabel('Phase Space', fontsize=12, fontweight='bold')
ax4.set_title('Energy Heatmap in Parameter Space', fontsize=13, fontweight='bold')
cbar4 = fig3.colorbar(im, ax=ax4)
cbar4.set_label('Energy (Hartree)', fontsize=10, fontweight='bold')
ax4.legend(fontsize=10, loc='upper right')

plt.tight_layout()
plt.savefig('vqe_3d_visualization.png', dpi=300, bbox_inches='tight',
            facecolor='white')
print("✓ 3D visualization saved as 'vqe_3d_visualization.png'")
plt.show()

# ============================================================================
# PLOT 4: Quantum State Visualization (Bloch Sphere Style)
# ============================================================================
fig4, axes = plt.subplots(1, 3, figsize=(18, 6))

# Sample three different theta values
theta_samples = [0, optimized_theta, np.pi]
titles = ['Initial State (θ=0)',
          f'Optimal State (θ={optimized_theta:.3f})',
          'Reference State (θ=π)']

for idx, (theta_sample, title) in enumerate(zip(theta_samples, titles)):
    ax = axes[idx]

    # Get state vector
    circuit = create_ansatz(theta_sample)
    result = simulator.simulate(circuit)
    state_vector = result.final_state_vector

    # Calculate probabilities
    probs = np.abs(state_vector)**2
    states = ['|00⟩', '|01⟩', '|10⟩', '|11⟩']

    # Create colorful bar chart
    colors_states = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']
    bars = ax.bar(states, probs, color=colors_states, alpha=0.8,
                  edgecolor='black', linewidth=2)

    # Add probability labels
    for bar, prob in zip(bars, probs):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                f'{prob:.3f}', ha='center', va='bottom',
                fontweight='bold', fontsize=11)

    ax.set_ylabel('Probability', fontsize=12, fontweight='bold')
    ax.set_title(title, fontsize=13, fontweight='bold', pad=10)
    ax.set_ylim([0, 1.0])
    ax.grid(True, alpha=0.3, axis='y')
    ax.set_xlabel('Basis States', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig('vqe_quantum_states.png', dpi=300, bbox_inches='tight',
            facecolor='white')
print("✓ Quantum state visualization saved as 'vqe_quantum_states.png'")
plt.show()

# ============================================================================
# PLOT 5: Interactive 3D Rotating Energy Surface
# ============================================================================
print("\n" + "=" * 70)
print("GENERATING INTERACTIVE 3D ROTATION SIMULATION")
print("=" * 70)

from matplotlib.animation import FuncAnimation, PillowWriter

# Create high-resolution 3D surface data
theta_3d = np.linspace(0, 2 * np.pi, 100)
phi_3d = np.linspace(0, np.pi, 50)
THETA, PHI = np.meshgrid(theta_3d, phi_3d)

# Create energy surface based on theta
Z_energy = np.zeros_like(THETA)
for i in range(THETA.shape[0]):
    for j in range(THETA.shape[1]):
        Z_energy[i, j] = compute_energy(THETA[i, j])

# Convert to Cartesian coordinates for 3D plotting
X_surf = np.sin(PHI) * np.cos(THETA)
Y_surf = np.sin(PHI) * np.sin(THETA)
Z_surf = Z_energy

fig5 = plt.figure(figsize=(12, 10))
ax_3d = fig5.add_subplot(111, projection='3d')

# Create the surface
surf = ax_3d.plot_surface(THETA, PHI, Z_energy, cmap='turbo',
                          alpha=0.9, edgecolor='none',
                          antialiased=True, linewidth=0)

# Add contour projections
ax_3d.contour(THETA, PHI, Z_energy, levels=15,
              cmap='turbo', linestyles='solid',
              offset=Z_energy.min()-0.1, alpha=0.6)

# Mark the optimal point
optimal_phi = np.pi / 2  # Middle of phi range
ax_3d.scatter([optimized_theta], [optimal_phi], [min_energy],
              color='red', s=300, marker='*',
              edgecolors='yellow', linewidths=3,
              label='Global Minimum', zorder=10)

# Plot the energy curve at phi = pi/2
energy_curve = [compute_energy(t) for t in theta_3d]
ax_3d.plot(theta_3d, [optimal_phi] * len(theta_3d), energy_curve,
           'r-', linewidth=4, alpha=0.8, label='Energy Landscape')

# Styling
ax_3d.set_xlabel('θ (radians)', fontsize=13, fontweight='bold', labelpad=12)
ax_3d.set_ylabel('φ (phase)', fontsize=13, fontweight='bold', labelpad=12)
ax_3d.set_zlabel('Energy (Hartree)', fontsize=13, fontweight='bold', labelpad=12)
ax_3d.set_title('3D Energy Surface with Optimal Point',
                fontsize=15, fontweight='bold', pad=20)
ax_3d.legend(loc='upper right', fontsize=11)

# Add colorbar
cbar = fig5.colorbar(surf, ax=ax_3d, shrink=0.6, aspect=10, pad=0.1)
cbar.set_label('Energy (Hartree)', fontsize=12, fontweight='bold')

# Set optimal viewing angle
ax_3d.view_init(elev=25, azim=45)

plt.tight_layout()
plt.savefig('vqe_3d_surface.png', dpi=300, bbox_inches='tight',
            facecolor='white')
print("✓ 3D energy surface saved as 'vqe_3d_surface.png'")
plt.show()

# ============================================================================
# Create Rotating Animation
# ============================================================================
print("\nGenerating 3D rotation animation (this may take a moment)...")

fig_anim = plt.figure(figsize=(12, 10))
ax_anim = fig_anim.add_subplot(111, projection='3d')

def init_animation():
    """Initialize the animation"""
    ax_anim.clear()
    return []

def animate(frame):
    """Animation function - rotates the 3D plot"""
    ax_anim.clear()

    # Recreate surface
    surf_anim = ax_anim.plot_surface(THETA, PHI, Z_energy, cmap='turbo',
                                     alpha=0.9, edgecolor='none',
                                     antialiased=True)

    # Add contours
    ax_anim.contour(THETA, PHI, Z_energy, levels=15,
                    cmap='turbo', linestyles='solid',
                    offset=Z_energy.min()-0.1, alpha=0.6)

    # Mark optimal point
    ax_anim.scatter([optimized_theta], [optimal_phi], [min_energy],
                    color='red', s=300, marker='*',
                    edgecolors='yellow', linewidths=3, zorder=10)

    # Energy curve
    ax_anim.plot(theta_3d, [optimal_phi] * len(theta_3d), energy_curve,
                 'r-', linewidth=4, alpha=0.8)

    # Labels
    ax_anim.set_xlabel('θ (radians)', fontsize=12, fontweight='bold', labelpad=10)
    ax_anim.set_ylabel('φ (phase)', fontsize=12, fontweight='bold', labelpad=10)
    ax_anim.set_zlabel('Energy (Hartree)', fontsize=12, fontweight='bold', labelpad=10)
    ax_anim.set_title(f'3D Energy Surface (Rotation: {frame}°)',
                      fontsize=14, fontweight='bold', pad=15)

    # Rotate view
    ax_anim.view_init(elev=20 + 10*np.sin(frame*np.pi/180), azim=frame)

    return []

# Create animation (360 degree rotation)
anim = FuncAnimation(fig_anim, animate, init_func=init_animation,
                     frames=np.linspace(0, 360, 120),
                     interval=50, blit=False)

# Save animation as GIF
writer = PillowWriter(fps=20)
anim.save('vqe_3d_rotation.gif', writer=writer, dpi=100)
print("✓ 3D rotation animation saved as 'vqe_3d_rotation.gif'")
plt.close(fig_anim)

# ============================================================================
# PLOT 6: Molecular Orbital Visualization
# ============================================================================
print("\nGenerating molecular orbital visualization...")

fig6 = plt.figure(figsize=(16, 10))

# Create 3D molecular structure visualization
ax_mol = fig6.add_subplot(221, projection='3d')

# H2 molecule positions (atomic units)
bond_length = 1.4  # Approximate bond length in Bohr
H1_pos = np.array([-bond_length/2, 0, 0])
H2_pos = np.array([bond_length/2, 0, 0])

# Draw atoms
ax_mol.scatter(*H1_pos, s=1000, c='cyan', alpha=0.8,
               edgecolors='blue', linewidths=2, label='H₁')
ax_mol.scatter(*H2_pos, s=1000, c='cyan', alpha=0.8,
               edgecolors='blue', linewidths=2, label='H₂')

# Draw bond
ax_mol.plot([H1_pos[0], H2_pos[0]],
            [H1_pos[1], H2_pos[1]],
            [H1_pos[2], H2_pos[2]],
            'k-', linewidth=4, alpha=0.7)

# Create electron cloud visualization (bonding orbital)
theta_cloud = np.linspace(0, 2*np.pi, 30)
phi_cloud = np.linspace(0, np.pi, 20)
THETA_cloud, PHI_cloud = np.meshgrid(theta_cloud, phi_cloud)

# Molecular orbital (simplified)
for center in [H1_pos, H2_pos]:
    r = 0.8
    X_cloud = center[0] + r * np.sin(PHI_cloud) * np.cos(THETA_cloud)
    Y_cloud = center[1] + r * np.sin(PHI_cloud) * np.sin(THETA_cloud)
    Z_cloud = center[2] + r * np.cos(PHI_cloud)

    ax_mol.plot_surface(X_cloud, Y_cloud, Z_cloud,
                        color='red', alpha=0.2, edgecolor='none')

ax_mol.set_xlabel('X (Bohr)', fontweight='bold')
ax_mol.set_ylabel('Y (Bohr)', fontweight='bold')
ax_mol.set_zlabel('Z (Bohr)', fontweight='bold')
ax_mol.set_title('H₂ Molecular Structure', fontsize=13, fontweight='bold')
ax_mol.legend()
ax_mol.set_box_aspect([1,1,1])

# Qubit State Vector Visualization
ax_qubits = fig6.add_subplot(222, projection='3d')

# Get optimal state vector
circuit_opt = create_ansatz(optimized_theta)
result_opt = simulator.simulate(circuit_opt)
state_opt = result_opt.final_state_vector

# Plot amplitudes as 3D bars
states_labels = ['|00⟩', '|01⟩', '|10⟩', '|11⟩']
x_pos = np.arange(len(states_labels))
y_pos = np.zeros(len(states_labels))
z_pos = np.zeros(len(states_labels))

real_parts = np.real(state_opt)
imag_parts = np.imag(state_opt)
magnitudes = np.abs(state_opt)

colors_bars = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']

for i in range(len(states_labels)):
    ax_qubits.bar3d(x_pos[i], 0, 0, 0.6, 0.6, magnitudes[i],
                    color=colors_bars[i], alpha=0.8,
                    edgecolor='black', linewidth=1.5)

ax_qubits.set_xlabel('Basis State', fontweight='bold')
ax_qubits.set_xticks(x_pos + 0.3)
ax_qubits.set_xticklabels(states_labels)
ax_qubits.set_ylabel('', fontweight='bold')
ax_qubits.set_zlabel('Amplitude', fontweight='bold')
ax_qubits.set_title('Optimal Quantum State Amplitudes',
                    fontsize=13, fontweight='bold')
ax_qubits.view_init(elev=20, azim=45)

# Energy Level Diagram (3D)
ax_energy = fig6.add_subplot(223, projection='3d')

# Create energy levels
energy_levels = [min_energy, min_energy + 0.2, min_energy + 0.5]
level_labels = ['Ground State', '1st Excited', '2nd Excited']
colors_levels = ['#FF6B6B', '#FFD93D', '#6BCB77']

for i, (energy, label, color) in enumerate(zip(energy_levels, level_labels, colors_levels)):
    # Draw energy level as a plane
    x_level = np.linspace(-1, 1, 10)
    y_level = np.linspace(-1, 1, 10)
    X_level, Y_level = np.meshgrid(x_level, y_level)
    Z_level = np.ones_like(X_level) * energy

    ax_energy.plot_surface(X_level, Y_level, Z_level,
                          color=color, alpha=0.6,
                          edgecolor='black', linewidth=0.5)

    # Add label
    ax_energy.text(0, 0, energy, label, fontsize=10,
                   fontweight='bold', ha='center')

ax_energy.set_xlabel('X', fontweight='bold')
ax_energy.set_ylabel('Y', fontweight='bold')
ax_energy.set_zlabel('Energy (Hartree)', fontweight='bold')
ax_energy.set_title('Energy Level Diagram', fontsize=13, fontweight='bold')
ax_energy.view_init(elev=15, azim=45)

# Optimization Trajectory in 3D
ax_traj = fig6.add_subplot(224, projection='3d')

# Create 3D optimization path
theta_opt_path = np.linspace(initial_theta[0], optimized_theta, 50)
energy_opt_path = [compute_energy(t) for t in theta_opt_path]
time_steps = np.linspace(0, 1, 50)

# Plot trajectory
for i in range(len(theta_opt_path)-1):
    ax_traj.plot([theta_opt_path[i], theta_opt_path[i+1]],
                 [time_steps[i], time_steps[i+1]],
                 [energy_opt_path[i], energy_opt_path[i+1]],
                 'b-', linewidth=2, alpha=0.6)

# Color gradient along path
scatter_traj = ax_traj.scatter(theta_opt_path, time_steps, energy_opt_path,
                               c=energy_opt_path, cmap='plasma',
                               s=80, alpha=0.8, edgecolors='black', linewidth=1)

# Mark start and end
ax_traj.scatter(theta_opt_path[0], time_steps[0], energy_opt_path[0],
                color='lime', s=300, marker='o',
                edgecolors='black', linewidth=2, label='Start', zorder=5)
ax_traj.scatter(theta_opt_path[-1], time_steps[-1], energy_opt_path[-1],
                color='red', s=400, marker='*',
                edgecolors='yellow', linewidth=3, label='Minimum', zorder=5)

ax_traj.set_xlabel('θ (radians)', fontweight='bold')
ax_traj.set_ylabel('Optimization Progress', fontweight='bold')
ax_traj.set_zlabel('Energy (Hartree)', fontweight='bold')
ax_traj.set_title('3D Optimization Trajectory', fontsize=13, fontweight='bold')
ax_traj.legend()
fig6.colorbar(scatter_traj, ax=ax_traj, label='Energy')
ax_traj.view_init(elev=20, azim=135)

plt.tight_layout()
plt.savefig('vqe_molecular_orbital_3d.png', dpi=300, bbox_inches='tight',
            facecolor='white')
print("✓ Molecular orbital 3D visualization saved as 'vqe_molecular_orbital_3d.png'")
plt.show()

print("\n" + "=" * 70)
print("ALL VISUALIZATIONS GENERATED SUCCESSFULLY!")
print("=" * 70)
print("\nGenerated files:")
print("  1. vqe_energy_landscape.png         - Enhanced 2D energy landscape")
print("  2. vqe_energy_contributions.png     - Hamiltonian term breakdown")
print("  3. vqe_3d_visualization.png         - 4-panel 3D analysis")
print("  4. vqe_quantum_states.png           - Quantum state probabilities")
print("  5. vqe_3d_surface.png               - Interactive 3D energy surface")
print("  6. vqe_3d_rotation.gif              - Rotating 3D animation (GIF)")
print("  7. vqe_molecular_orbital_3d.png     - Molecular orbital 4-panel 3D")

print("\n" + "=" * 70)
print("VQE COMPUTATION COMPLETE")
print("=" * 70)
```
